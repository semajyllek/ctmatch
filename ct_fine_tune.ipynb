```python
from pathlib import Path
TREC_DATA_PATH = Path('/content/drive/MyDrive/ct_data23/trec_data.jsonl')
KZ_DATA_PATH = Path('/content/drive/MyDrive/ct_data23/kz_data.jsonl')
```

```python
from ctmatch.match import ModelConfig, CTMatch
```

```python
model_config = ModelConfig(
  data_path=KZ_DATA_PATH,
  model_checkpoint='allenai/scibert_scivocab_uncased',
  max_length=512,
  batch_size=16,
  padding='max_length',
  truncation=True,
  learning_rate=2e-5,
  num_train_epochs=5,
  weight_decay=0.01,
  warmup_steps=500,
  seed=37,
  splits={"train":0.8, "val":0.1},
  output_dir = "/content/drive/MyDrive/scibert_finetuned_ct",
)

ctm = CTMatch(model_config)
```

```python
ctm.ct_dataset
```

```python
def train_and_predict(model, dataset):
  model.train()
  predictions = model.predict(dataset["test"])
  return predictions
```

```python
train_and_predict(ctm.model, ctm.ct_dataset)
```
